```python
import random

def roll_dice():
    return random.randint(1, 6)

def movement(player, steps):
    player += steps
    if player > 100:
        player = 100
    return player

def snake_or_ladder(player):
    snakes = {17: 7, 54: 34, 62: 19, 64: 60, 87: 24, 93: 73, 95: 75, 98: 79}
    ladders = {1: 38, 4: 14, 9: 31, 21: 42, 28: 84, 51: 67, 71: 91, 80: 100}
    if player in snakes:
        print("Oops! You landed on a snake. Go back to position", snakes[player])
        return snakes[player]
    elif player in ladders:
        print("Yay! You found a ladder. Climb up to position", ladders[player])
        return ladders[player]
    else:
        return player
    
def input_player(player): 
    n=input("Press Enter to roll the dice...")
    if(n==""):
        dice = roll_dice()
        print("You rolled:", int(dice) )
        player = movement(player, int(dice))
        player = snake_or_ladder(player)
        print("You are now at position", player)
        if dice == 6:
            print("You get another turn!")
            print("\n")
            input_player(player)
    return player


def main():
    player1 = 0
    player2 = 0
    while True:
        print("-----Player1-----")
        print("\n")
        player1 = input_player(player1)
        if player1 == 100:
            print("Player 1 wins!")
            break
        
        print("\n")
        print("---------------------------------")
        print("\n")
        
        print("------Player2------")
        print("\n")
        player2 = input_player(player2)
        if player2 == 100:
            print("Player 2 wins!")
            break


main()
```

Algorithm for the snake and ladder :- 
1. Define functions:
   - roll_dice()
   - movement(player, steps)
   - snake_or_ladder(player)
   - input_player(player)

2. Initialize player positions:
   - player1 = 0
   - player2 = 0

3. Game Loop:
   while True:
       For each player in [player1, player2]:
           Prompt the player to press Enter to roll the dice.
           If the player presses Enter:
               Roll the dice using roll_dice().
               Print the dice roll result.
               Move the player's position based on the dice roll using movement().
               Check if the player lands on a snake or ladder using snake_or_ladder().
               Print the new player position.
               If the player rolls a 6, they get another turn.
           If the player reaches position 100:
               Print that the player wins.
               Break out of the loop.

4. Main Function:
   Call the main function to start the game loop.
